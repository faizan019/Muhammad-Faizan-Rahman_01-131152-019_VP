```python
from google.cloud import bigquery
```

```python
client = bigquery.Client()
```

```python
# Set up the data to be streamed
data = [
    ('2022-12-13T10:45:00', 'sp_zenith', 151515.9288238668 ),
    ('2022-12-13T10:45:00', 'air_temperature', 202020.9717641772)
]
```

```python
# Set up the table and fields in BigQuery
table_id = 'outstanding-pen-377709.faizan_dataset.my_gbq_table010'
schema = [
    {"name":"time_local", "type":"DATETIME"},
    {"name": "alias", "type": "STRING"},
    {"name":"value", "type": "FLOAT"}
]
```

```python
table = bigquery.Table(table_id, schema=schema)
```

```python
# Stream the data into BigQuery
stream_data = client.insert_rows(table, data)
```

```python
if stream_data == []:
    print('Data streamed successfully into BigQuery.')
else:
    print(f'Error streaming data into BigQuery: {stream_data}')
```

```python
from google.cloud import storage
import pandas as pd
import io

def download_file_as_dataframe(bucket_name, file_name):
    # Create a GCS client object
    storage_client = storage.Client()

    # Get the GCS bucket object
    bucket = storage_client.bucket(bucket_name)

    # Get the GCS blob object for the file
    blob = bucket.blob(file_name)

    # Download the contents of the file as a byte string
    file_contents = blob.download_as_string()

    # Convert the byte string to a pandas dataframe
    df = pd.read_csv(io.StringIO(file_contents.decode('utf-8')))
    #df = pd.read_csv(blob.download_as_text())

    # Return the pandas dataframe
    return df
```
