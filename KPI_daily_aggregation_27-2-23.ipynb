```python
#3-using apply function
import pandas as pd
import numpy as np

start_date = '2023-01-01 00:00:00'
end_date = '2023-01-05 00:00:00'

dates = pd.date_range(start=start_date, end=end_date, freq='15min')
values = np.random.randn(len(dates))

df = pd.DataFrame({'date': dates, 'point': values})

daily_min_max = df.groupby(pd.Grouper(key='date', freq='D')).agg({'point': ['min', 'max']})
daily_min_max['Value'] = daily_min_max.apply(lambda x: x[('point', 'max')] - x[('point', 'min')], axis=1)

eq = 'x[("point", "max")] - x[("point", "min")]'
def func(x):
    y = eval(eq)
    return y

daily_min_max['Value1'] = daily_min_max.apply(func, axis=1)

print(daily_min_max)
```

```python
#5-first_value and last_value using eval()
import pandas as pd
import numpy as np

start_date = '2023-01-01 00:00:00'
end_date = '2023-01-03 00:00:00'

dates = pd.date_range(start=start_date, end=end_date, freq='15min')
values = np.random.randn(len(dates))

df = pd.DataFrame({'date': dates, 'rand_num': values})

df['first_value'] = df.groupby(df.date.dt.date)['rand_num'].transform('first')
df['last_value'] = df.groupby(df.date.dt.date)['rand_num'].transform('last')

#eq = 'max(values) - min(values)'
eq= 'last_value - first_value'

#df['value'] = df.groupby(df.date.dt.date).apply(lambda x: eval(eq), axis=1)
df['value'] = df.eval(eq)

print(df)
```
