```python

def extract_params(func_call):
    pattern = r'^(\w+)\((.+)\)$'
    match = re.match(pattern, func_call)
    if not match:
        raise ValueError("Invalid function call")
    
    params_str = match.group(2)
    params = {}
    
    for param in params_str.split(','):
        param = param.strip()
        param_match = re.match(r'^(\w+)\s*=\s*[\'"]?(\w+)[\'"]?$', param)
        
        if not param_match:
            raise ValueError("Invalid parameter")
        
        param_name = param_match.group(1)
        param_value = param_match.group(2)
        
        params[param_name] = param_value
#         params_list = list(params.items())
    return params
```

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows',6000)

df = pd.DataFrame()
df['time'] = pd.date_range(start='2023-01-31',end='2023-03-24 05:00:00',freq='15min')
df['value'] = np.random.randint(1, 15, size=len(df))

a = 270.5643
constant = -6325.637

# Resampling to daily frequency and calculating the mean of the 'value' column
daily_mean = df.set_index('time').resample('D')['value'].mean()

# Applying the daily mean to each 15-minute interval using 'transform'
df['daily_mean'] = df.groupby(df['time'].dt.date)['value'].transform('mean')

# Calculating the 'equation' column using the daily mean, a, and constant
df['equation'] = ((df['daily_mean'] * a) / constant)

print(df)
```

```python
import pandas as pd
import numpy as np

# pd.set_option('display.max_rows',2000)

df = pd.DataFrame()
df['time'] = pd.date_range(start='2023-01-31',end='2024-03-05',freq='15min')
df['value'] = np.random.randint(1, 15, size=len(df))

a = 270.5643
constant = -6325.637

# Calculating the daily mean of the "value" column
daily_mean = df.resample('D', on='time').mean()

# Adding the "equation" column to the daily mean df
daily_mean['equation'] = (daily_mean['value'] * a) - constant

daily_mean['year_month'] = daily_mean.index.strftime('%Y-%m')
# print(daily_mean)

# Group by year_month and count the number of days 
days_counts = daily_mean.groupby('year_month').agg({'value': 'count'})

# Rename the 'value' column to 'days_in_month'
days_counts = days_counts.rename(columns={'value': 'days_in_month'})
# print(days_counts)

# Join the two dataframes based on the year_month column
final_df = daily_mean.merge(days_counts, on='year_month')

# Create a new column called equation_per_day
final_df['equation_per_day'] = final_df['equation'] / final_df['days_in_month']

# Display the merged dataframe with the new column
display(final_df)
```
