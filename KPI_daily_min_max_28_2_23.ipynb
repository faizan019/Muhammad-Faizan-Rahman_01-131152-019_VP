```python
#This script is to find the range from the last and the first value for each day through the equation 

import pandas as pd
import numpy as np


start_date = '2023-01-01 00:00:00'
end_date = '2023-01-05 00:00:00'

dates = pd.date_range(start=start_date, end=end_date, freq='15min')
values = np.random.randn(len(dates))

df = pd.DataFrame({'date': dates, 'Values': values})
df.set_index('date', inplace = True)

df = df.resample('D')['Values'].agg(['first','last']).reset_index()

equation = "df['last'] - df['first']"
df['value'] = eval(equation)

df
```

```python
#Through max and min 

start_date = '2023-01-01 00:00:00'
end_date = '2023-01-10 00:00:00'

dates = pd.date_range(start=start_date, end=end_date, freq='15min')
values = np.random.randn(len(dates))

df1 = pd.DataFrame({'date': dates, 'Values': values})
df1= df1.groupby(pd.Grouper(key='date', freq='D')).agg({'Values': ['min','max']}).reset_index()
#print(df1.columns)

df1.columns = ['_'.join(col).strip() for col in df1.columns.values]
#print(df1.columns)

equation = "df1['Values_max'] - df1['Values_min']"
df1['Value'] = eval(equation)

df1
```
