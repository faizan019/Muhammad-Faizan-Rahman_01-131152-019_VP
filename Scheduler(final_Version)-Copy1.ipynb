```python
def get_current_value(schedule, time_now):
    
    current_day = time_now.date()
    current_time = time_now.time()
    if current_day in schedule:
        for interval in schedule[current_day]:
            start_time = datetime.combine(current_day, time(hour=interval["hour"], minute=interval["minute"], second=interval["second"]))
            if current_time >= start_time:
                # if current time is within the interval, return the corresponding value
                return interval["value"]
           

            elif current_time == time(hour=23, minute=59, second=59) and interval.get("hour") == 23 and interval.get("minute") == 59 and interval.get("second") == 59:
                return interval["value"]
    return None
```
